```julia
using LinearAlgebra
using Plots
```

```julia
X = 10.0
I = 500
δx = X/I
xs = [i*δx for i in -I:I]
N = length(xs);
```

setter opp et romlig gitter for systemet. $N$ er størrelsen på gitteret.

```julia
L = -5.0
ΔL = 1.0
∇L = 140.0

R = 5.0
ΔR = 1.0
∇R = 120.0

Π(x,A,ΔA,∇A) = -∇A*exp(-(x-A)^2/2ΔA)
ΠL(x) = Π(x,L,ΔL,∇L)
ΠR(x) = Π(x,R,ΔR,∇R)

ΠRs = [ΠL(x) for x in xs];
ΠLs = [ΠR(x) for x in xs];
Πs = ΠRs+ΠLs;
```

spesifiserer brønnenes posisjon, bredde og dybde og potergifunksjon.

```julia
HL = SymTridiagonal(ΠLs+repeat([1/δx^2],N),repeat([-1/δx^2],N-1));
HR = SymTridiagonal(ΠRs+repeat([1/δx^2],N),repeat([-1/δx^2],N-1));
eigenL = eigen(HL); ΨLs = eigenL.vectors; ELs = eigenL.values;
eigenR = eigen(HR); ΨRs = eigenR.vectors; ERs = eigenR.values;
```

```julia
plot(xs,ΨLs[:,1];color="#fdce0b")
plot!(xs,ΨRs[:,1];color="#fdce0b")
```
