```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from sklearn.metrics import mean_squared_error as mse
matplotlib.rcParams.update({'font.size': 12})
```

## 2D Plate

```python
def u(x, y, t):
    return np.sin(np.pi * x) * np.sin(np.pi * y) * np.exp(-2 * np.pi**2 * t)
```

```python
times = [0, 0.01, 0.05]
```

```python
for scheme, name in zip(["imp", "exp"], ["Implicit", "Explicit"]):
    for n in [11, 101]:
        df = pd.read_csv(f"Output/plate_{scheme}_{n}.txt", delim_whitespace=True, header=None).transpose()
        df.columns = df.iloc[0]
        df = df[1:]

        dx = 1 / (n - 1)
        X = np.linspace(0, 1, n)
        X, Y = np.meshgrid(X, X)

        for ti, time in zip(df, times):
            plt.figure(figsize=(8,6))
            z = np.array(df[ti]).reshape((n,n))
            print(f"MSE = {mse(z, u(X, Y, time)):.3e}")
            plt.title(f"{name}: dx = {dx}, t = {time}")
            plt.contourf(X, Y, z, 25, cmap='RdGy')
            plt.colorbar()
            plt.clim(0, 1)
            plt.show()
```
