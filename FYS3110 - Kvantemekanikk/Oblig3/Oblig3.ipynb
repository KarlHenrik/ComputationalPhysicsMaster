```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
V = 5
g = 1

L = 25
#Hamiltonian
H = np.zeros((L, L))
H[0, 0] = -V
H[0, 1] = -g
for i in range(1, L-1):
    H[i, i-1] = -g
    H[i, i+1] = -g
H[L-1, L-2] = -g
#X
X = np.zeros((L, L))
for i in range(L):
    X[i, i] = i
```

```python
eigVals = np.linalg.eig(H)[0]
eigVecs = np.linalg.eig(H)[1]
plt.hlines(eigVals, 0, 1)
plt.xticks([])
plt.show()
```

```python
GS = eigVecs[:,0]
print(GS[0]**2)
```

```python
eigVecs[:,0][1]**2
```

```python
np.inner(GS, X @ GS)
```

```python
v0 = np.zeros(L); v0[0] = 1
coeffs = np.linalg.solve(eigVecs, v0)
hbar = 1
def prob0(t):
    total = 0
    for i in range(L):
        total += eigVecs[:,i][0] * coeffs[i] * np.exp(-1j / hbar * eigVals[i] * t)
    return np.abs(total)**2
```

```python
t = np.linspace(0, 100, 1000)
plt.plot(t, prob0(t))
plt.show()
```
