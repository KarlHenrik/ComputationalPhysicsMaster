## 1D constant gravity

```julia
function ec1(pos, vel, acc, N, dt, grav)
    for i in 1:N-1
    acc[i] = grav(pos[i])
    vel[i + 1] = vel[i] + acc[i] * dt
    pos[i + 1] = pos[i] + vel[i] * dt
    end
    return pos, vel, acc
end;
```

```julia
pos_0 = 0
vel_0 = 100
N = 100000
dt = 20 / N
g = 9.81

pos = zeros(N)
vel = zeros(N)
acc = zeros(N)

pos[1] = pos_0
vel[1] = vel_0

grav(pos) = -g;

pos, vel, acc = ec1(pos, vel, acc, N, dt, grav)
plt.plot(pos, line = 1.5)
```

## 2D Gravity

```julia
pos_0 = [0, 1000] # moon average distance to Earth in km
vel_0 = [130, 0] # moon average velocity in km/s
N = 100000
dt = 0.1
GM = 1e7

pos = zeros(2, N)
vel = zeros(2, N)
acc = zeros(2, N)

pos[:, 1] = pos_0
vel[:, 1] = vel_0;

grav(pos) = -pos .* GM ./ la.norm(pos)^3

for i in 1:N-1
    acc[:, i] = grav(pos[:, i])
    vel[:, i + 1] = @. vel[:, i] + acc[:, i] * dt
    pos[:, i + 1] = @. pos[:, i] + vel[:, i + 1] * dt
end

plt.plot(pos[1, :], pos[2, :], line = 3)

plt.plot!(xlims = (-4000, 4000), ylims = (-6000, 2000), size = (400, 400))
```
