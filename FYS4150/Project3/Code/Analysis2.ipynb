```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
nameMass = np.array(pd.read_csv("Output/test1.txt", header=None, delim_whitespace=True, nrows=1))[0]

df = pd.read_csv("Output/test1.txt", header=None, delim_whitespace=True, skiprows = 1)
```

```python
n_planets = len(nameMass) // 2
names = [nameMass[2*i] for i in range(n_planets)]
masses = [nameMass[2*i + 1] for i in range(n_planets)]
```

```python
plt.figure(figsize=(8,8))
for i in range(n_planets):
    plt.plot(df[6*i], df[6*i + 1], label=names[i])
plt.show()
```

```python
plt.figure(figsize=(8,8))
for i in [0,1,3,4]:
    plt.plot(df[6*i], df[6*i + 1], label=names[i])
plt.legend()
plt.show()
```
