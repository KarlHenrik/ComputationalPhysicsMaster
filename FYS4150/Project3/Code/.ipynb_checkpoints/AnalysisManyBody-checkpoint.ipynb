# Many Body Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Earth and Jupyter with stationary sun

```python
fileloc = "Output/jupyterMass"
for filename, jmass in zip(["1.000000.txt", "10.000000.txt", "1000.000000.txt"], ["1", "10", "1000"]):
    plt.figure(figsize=(5, 5))
    nameMass = np.array(pd.read_csv(fileloc + filename, header=None, delim_whitespace=True, nrows=1))[0]

    df = pd.read_csv(fileloc + filename, header=None, delim_whitespace=True, skiprows = 1)
    n_planets = df.shape[1] // 3
    names = [nameMass[2*i] for i in range(n_planets)]
    masses = [nameMass[2*i + 1] for i in range(n_planets)]

    for i in range(n_planets):
        plt.plot(df[3*i], df[3*i + 1], label=names[i])
    plt.legend()
    plt.xlim(-6, 6)
    plt.ylim(-6, 6)
    plt.xlabel("x (AU)")
    plt.ylabel("y (AU)")
    plt.title(f"Baner med Jupyter som har {jmass} ganger sin vanlige masse")
    plt.show()
```

## Earth and Jupyter with moving sun

```python
fileloc = "Output/jupyterMassMove"
for filename, jmass in zip(["1.000000.txt", "10.000000.txt", "1000.000000.txt"], ["1", "10", "1000"]):
    plt.figure(figsize=(5, 5))
    nameMass = np.array(pd.read_csv(fileloc + filename, header=None, delim_whitespace=True, nrows=1))[0]

    df = pd.read_csv(fileloc + filename, header=None, delim_whitespace=True, skiprows = 1)
    n_planets = df.shape[1] // 3
    names = [nameMass[2*i] for i in range(n_planets)]
    masses = [nameMass[2*i + 1] for i in range(n_planets)]

    for i in range(n_planets):
        plt.plot(df[3*i], df[3*i + 1], label=names[i])
    plt.legend()
    plt.xlim(-6, 6)
    plt.ylim(-6, 6)
    plt.xlabel("x (AU)")
    plt.ylabel("y (AU)")
    plt.title(f"Baner med Jupyter som har {jmass} ganger sin vanlige masse")
    plt.show()
```

## All planets

```python
filename = "Output/allplong.txt"
plt.figure(figsize=(5, 5))
nameMass = np.array(pd.read_csv(filename, header=None, delim_whitespace=True, nrows=1))[0]
df = pd.read_csv(filename, header=None, delim_whitespace=True, skiprows = 1)
n_planets = df.shape[1] // 3

names = [nameMass[2*i] for i in range(n_planets)]
masses = [nameMass[2*i + 1] for i in range(n_planets)]

for i in range(n_planets):
    plt.plot(df[3*i], df[3*i + 1], label=names[i])
plt.legend(loc = "upper right")
plt.xlabel("x (AU)")
plt.ylabel("y (AU)")
plt.show()
```

## How all of this affects the Earth

```python
def plotEarth(filename, index, label):
    df = pd.read_csv(filename, header=None, delim_whitespace=True, skiprows = 1)
    n_planets = df.shape[1] // 3

    plt.plot(df[index], df[index + 1], label = label)
```

```python
plt.figure(figsize=(5, 5))

plotEarth("Output/allplong.txt", 9, "All planets")
plotEarth("Output/onlyEarth.txt", 0, "Only Earth")
plt.xlabel("x (AU)")
plt.ylabel("y (AU)")
plt.legend(loc = "upper right")
plt.show()
```
