```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
folderPath = "../Output/earthSun"
for i in [2, 3, 4, 5]:
    for solver, splot in zip(["euler", "vverlet"], [121, 122]):
        df = pd.read_csv(f"{folderPath}/earthSun_{solver}{i}.txt", delim_whitespace=True)

        x = np.array(df["x"])
        y = np.array(df["y"])
        z = np.array(df["z"])
        vx = np.array(df["vx"])
        vy = np.array(df["vy"])
        vz = np.array(df["vz"])

        plt.figure(figsize = (6, 6))
        if solver == "vverlet":
            solver = "Velocity verlet"
        plt.title(f"Orbit from {solver.capitalize()} with stepsize $10^{{{-i}}}$")
        plt.plot(x, y)
        
        plt.figure(figsize = (6, 4))
        plt.title(f"Kinetic and potential energy for {solver.capitalize()} with stepsize $10^{{{-i}}}$")
        plt.plot(-(4 * np.pi**2 * 3E-6) / (x**2 + y**2 + z**2)**0.5, label="Potential energy")
        plt.plot(0.5 * 3E-6 * (vx**2 + vy**2 + vz**2), label="Kinetic energy")
        plt.plot(-(4 * np.pi**2 * 3E-6) / (x**2 + y**2 + z**2)**0.5 + 0.5 * 3E-6 * (vx**2 + vy**2 + vz**2), label="Total energy")
        plt.legend()

    plt.show()
```
