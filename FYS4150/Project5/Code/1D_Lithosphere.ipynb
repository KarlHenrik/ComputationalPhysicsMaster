```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 12})
```

### Plotting time evolution of the Lithosphere

```python
times = [0, 0.01, 0.5, 1, 10]
```

```python
df = pd.read_csv("Output/lit_base.txt", delim_whitespace=True, header=None).transpose()
df.columns = df.iloc[0]
df = df[1:]
x = np.linspace(0, 1, len(df["t0"]))

for ui, time in zip(df, times):
    plt.plot(x * 120, df[ui]*1292+8, label=f"{time} Gy")
plt.legend()
plt.xlabel("Depth [km]")
plt.ylabel("Temperature [°C]")
plt.title("No enrichment")
plt.show()
```
